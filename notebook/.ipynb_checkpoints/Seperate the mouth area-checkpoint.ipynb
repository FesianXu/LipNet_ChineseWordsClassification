```python
'''
分离出嘴部的区域，使用alignment信息
'''
import numpy as np
import face_alignment
import cv2
import os
from skimage import io
from matplotlib import pyplot as plt
os.environ['CUDA_VISIBLE_DEVICES'] = '1'
%matplotlib inline
fa = face_alignment.FaceAlignment(face_alignment.LandmarksType._2D, flip_input=False)
```

```python
landmark_path = '/home/fesian/contest_workspace/chunyu/dataset/landmarks/'
images_path = '/home/fesian/contest_workspace/chunyu/dataset/'
tmode = ('train','test')
```

```python
mouth_bias_x = 20
mouth_bias_y = 20
for dataset in tmode:
    lpath = landmark_path+'{}data/'.format(dataset)
    ipath = images_path+'lip_{}/lip_{}/'.format(dataset,dataset)
    dirlist = os.listdir(lpath)
    for sample in dirlist:
        print('{} name'.format(sample))
        spath = lpath+sample
        landmark = np.load(spath)
        if landmark.shape[0] == 0: continue
        mouth_landmark = landmark[:, 48:68, :]
        img_path = ipath+'{}/'.format(sample[:-4])
        imgslist = os.listdir(img_path)
        for ind, imgsample in enumerate(imgslist):
            imgpath = img_path+imgsample
            imgmat = io.imread(imgpath)
            align = fa.get_landmarks(imgmat)
            if not align: continue
            align = align[0][48:68,:]
            max_x, max_y, min_x, min_y = np.max(align[:,0]), np.max(align[:,1]), np.min(align[:,0]), np.min(align[:,1])
            height, width, depth = imgmat.shape
            x1 = int(min_x)-mouth_bias_x if int(min_x)-mouth_bias_x >= 0 else 0
            x2 = int(max_x)+mouth_bias_x if int(max_x)+mouth_bias_x <= height-1 else height-1
            y1 = int(min_y)-mouth_bias_y if int(min_y)-mouth_bias_y >= 0 else 0
            y2 = int(max_y)+mouth_bias_y if int(max_y)+mouth_bias_y <= width-1 else width-1
            mouthimg = imgmat[int(min_y)-mouth_bias_y:int(max_y)+mouth_bias_y, int(min_x)-mouth_bias_x:int(max_x)+mouth_bias_x, :]
            if not os.path.exists(images_path+'mouth_{}/{}/'.format(dataset,sample[:-4])):
                os.makedirs(images_path+'mouth_{}/{}/'.format(dataset,sample[:-4]))
            save_path = images_path+'mouth_{}/{}/{}'.format(dataset,sample[:-4],imgsample)
            cv2.imwrite(save_path, mouthimg[:,:,::-1])
```
