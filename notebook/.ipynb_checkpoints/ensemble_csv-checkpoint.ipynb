```python
import numpy as np
import pandas as pd
import os
```

```python
rootpath = '/home/fesian/contest_workspace/chunyu/result/lipnet_result/'
predicts = [rootpath + x for x in os.listdir(rootpath)]
```

```python
dfs = []
for each in predicts:
    dfs += [pd.read_csv(each, header=None)]
```

```python
pps = {}
for eachp in dfs:
    for index, row in eachp.iterrows():
        hashid = row[0]
        pred = row[1]
        if hashid in pps: 
            pps[hashid] += [pred]
        else:
            pps[hashid] = []
```

```python
pps
```

```python
csv = []
for k, v in pps.items():
    count = {}
    for ev in v:
        if ev in count:
            count[ev] += 1
        else:
            count[ev] = 0
    p = max(count, key=count.get)
    csv.append((k, p))
```
