```python
# generate the character codebook instead of the word codebook
import numpy as np
```

```python
w_cb = np.load('/home/fesian/contest_workspace/chunyu/dataset/labels/codebook.npy').item()
```

```python
c_cb = {}
cnt = 0
for k, v in w_cb.items():
    for c in k:
        if c not in c_cb:
            c_cb[c] = cnt
            cnt += 1
```

```python
np.save('/home/fesian/contest_workspace/chunyu/dataset/ctc_labels/codebook.npy', c_cb)
```

```python
codebook =  {v: k for k, v in w_cb.items()}
```

```python
# 统计样本的字符长度
lengths = {}
```

```python
train_index = np.load('/home/fesian/contest_workspace/chunyu/dataset/labels/train_index.npy')
eval_index = np.load('/home/fesian/contest_workspace/chunyu/dataset/labels/eval_index.npy')
```

```python
train_index.shape
```

```python
for each in train_index:
    word = codebook[int(each[1])]
    lengths[each[0]] = (len(word), word, int(each[1]))
for each in eval_index:
    word = codebook[int(each[1])]
    lengths[each[0]] = (len(word), word, int(each[1]))
```

```python
np.save('/home/fesian/contest_workspace/chunyu/dataset/ctc_labels/train_eval_char_length.npy', lengths)
```
