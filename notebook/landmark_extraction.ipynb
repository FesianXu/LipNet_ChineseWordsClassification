```python
import face_alignment
from skimage import io
import cv2
from matplotlib import pyplot as plt
import os
import numpy as np
%matplotlib inline

fa = face_alignment.FaceAlignment(face_alignment.LandmarksType._2D, flip_input=False)
```

```python
samples_type = ['train', 'test']
rootpath = '/home/fesian/contest_workspace/chunyu/dataset/'
```

```python
for datasest in samples_type:
    datapath = rootpath+'lip_{}/lip_{}/'.format(datasest, datasest)
    datalist = os.listdir(datapath)
    missing_list = []
    for indsample, eachsample in enumerate(datalist):
        print('----{} dataset,  No. {} sample --- named {}'.format(datasest, indsample, eachsample))
        samplepath = datapath+eachsample+'/'
        imglist = os.listdir(samplepath)
        saved_landmark =  []
        paralist = [int(x[:-4]) for x in imglist]
        paralist.sort()
        paralist = ['{}.png'.format(x) for x in paralist]
        for img in paralist:
            imgpath = samplepath+img
            imgmat = io.imread(imgpath)
            preds = fa.get_landmarks(imgmat)
            if not preds:
                missing_list.append(
                    (datasest, eachsample, img)
                    )
                print(len(missing_list))
                continue
            preds = np.array({'pred': preds[0], 'id':img})
            saved_landmark.append(preds)
        saved_landmark = np.array(saved_landmark)
        save_path = rootpath+'landmarks_normal_dict/{}data/{}.npy'.format(datasest, eachsample)
        np.save(save_path, saved_landmark)
    missing_list = np.array(missing_list)
    np.save(rootpath+'landmarks_normal_dict/missing_img.npy', missing_list)
```
